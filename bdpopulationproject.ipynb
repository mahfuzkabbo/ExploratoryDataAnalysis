```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('city_population.csv')
```

```python
df
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.dropna()
```

```python
df.columns = df.columns.str.strip().str.replace(" ", "_")
df.head()
```

```python
numeric_cols = ['Area_(km2)','Population_1991','Population_2001',
                'Population_2011','Population_2022']

df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors='coerce')
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()

```

```python
df['Established'] = df['Established'].astype(int)
```

```python
df = df.drop(columns=['Native'])
```

```python
df = df.reset_index(drop=True)
```

```python
df
```

```python
df.head()
df.info()
df.describe()
```

```python
df['Division'].value_counts()
plt.figure(figsize=(10,5))
sns.countplot(data=df, x='Division')
plt.xticks(rotation=45)
plt.title("Number of Districts per Division")
plt.show()
```

```python
plt.figure(figsize=(14,5))
sns.barplot(data=df, x='Name', y='Area_(km2)')
plt.xticks(rotation=90)
plt.title("District Area Comparison")
plt.show()
```

```python
df['Pop_Growth_1991_2022'] = df['Population_2022'] - df['Population_1991']
```

```python
top_growth = df.nlargest(10, 'Pop_Growth_1991_2022')

plt.figure(figsize=(12,5))
sns.barplot(data=top_growth, x='Name', y='Pop_Growth_1991_2022')
plt.xticks(rotation=90)
plt.title("Top 10 Fastest Growing Districts (1991–2022)")
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.scatterplot(data=df, x='Area_(km2)', y='Population_2022')
plt.title("Area vs Population (2022)")
plt.show()
```

```python
dhaka = df[df['Name'] == 'Dhaka'].iloc[0]

years = [1991, 2001, 2011, 2022]
pop = [dhaka['Population_1991'],
       dhaka['Population_2001'],
       dhaka['Population_2011'],
       dhaka['Population_2022']]

plt.plot(years, pop, marker='o')
plt.title("Population Growth of Dhaka")
plt.xlabel("Year")
plt.ylabel("Population")
plt.show()
```
